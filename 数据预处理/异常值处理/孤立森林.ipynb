```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
sns.set()
```

```python
from sklearn.datasets import load_wine
dataset = load_wine()
X = dataset.data
y = dataset.target
```

```python
df = pd.DataFrame(X)
df.columns = dataset.feature_names
```

```python
df.head()
```

## 逐特征筛选

```python
idx = 1
for i in df.columns[:-2]:
    model=IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1), max_features=1.0)
    model.fit(df[[i]])
    df['scores'] = model.decision_function(df[[i]])
    df['anomaly'] = model.predict(df[[i]])

    tmp_lst = df['anomaly'] != -1
    idx &= tmp_lst
    # anomaly = df.loc[df['anomaly']==-1]
    # print(type(anomaly))
df_new = df[idx]
df_new
```

## 多特征一并处理

```python
lst = df.columns
model = IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1), max_features=1.0)
model.fit(df[lst])
df['scores'] = model.decision_function(df[lst])
df['anomaly'] = model.predict(df[lst])

anomaly = df.loc[df['anomaly']==-1]
print(len(anomaly))
print(len(df))
```
