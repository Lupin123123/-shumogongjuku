```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('final file.xlsx')
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
label = enc.fit_transform(df['cixing'])
df['cixing'] = label
df
```

```python
from sklearn.preprocessing import StandardScaler
enc = StandardScaler()

feature_list = df.columns[1:]
for i in feature_list:
    label = enc.fit_transform(df[i].to_numpy().reshape(-1, 1))
    df[i] = label
```

```python
df
```

```python
k = 7
def fun(a):
    s = 0
    for i in range(k+1, len(df)):
        x1 = df.iloc[i, 1]
        x2 = df.iloc[i, 2]
        x3 = df.iloc[i, 3]
        x4 = df.iloc[i, 4]
        x5 = df.iloc[i-1, 5]
        x6 = df.iloc[i-2, 5]
        x7 = df.iloc[i-3, 5]
        # x8 = df.iloc[i-4, 5]
        # x9 = df.iloc[i-5, 5]
        # x10 = df.iloc[i-6, 5]
        # x11 = df.iloc[i-7, 5]
        s += 0.5*(a[0]*x1 + a[1]*x2 + a[2]*x3 + a[3]*x4 + a[4]*x5 + a[5]*x6\
                  + a[6]*x7 \
                #   + a[7]*x8 + a[8]*x9 \
                  - df.iloc[i, -1])**2
    print(s)
    return s
```

```python
b_init = [0, 0, 0, 0, 0, 0, 0]
method_list = ['Nelder-Mead', 'Powell', 'CG', 'BFGS', 'L-BFGS-B']

res = minimize(fun=fun, x0=b_init, method='Nelder-Mead')
res
```
