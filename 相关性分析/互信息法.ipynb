```python
import pandas as pd
import numpy as np
from sklearn.metrics import normalized_mutual_info_score
# from sklearn.feature_selection import mutual_info_regression
```

```python
df = pd.read_csv('E:\workspace\数模工具库\dataset\\time_sequence\PRSA_data_2010.1.1-2014.12.31.csv')  
df.dropna(inplace=True)
```

```python
feature_lst = df.columns[5:]
dataDis = df.loc[:, feature_lst]                  # 需要归一化互信息处理的数据提取
```

```python
[_, varLenDis] = dataDis.shape
nmiMatrix = [[0 for col in range(varLenDis)] for row in range(varLenDis)]
nmiMatrix = pd.DataFrame(nmiMatrix, columns=feature_lst, index=feature_lst)
```

```python
for i in range(varLenDis):
    for j in range(varLenDis):
        nmiMatrix.iloc[i, j] = normalized_mutual_info_score(dataDis[feature_lst[i]].values, dataDis[feature_lst[j]].values, average_method='arithmetic')
print("**************离散变量相关性分析：互信息系数矩阵：*******************")
print(nmiMatrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.subplots(figsize=(9, 9))

plt.rcParams['font.sans-serif']=['FangSong']  # 正常显示文字
plt.rcParams['axes.unicode_minus']=False      # 正常显示负号

sns.heatmap(nmiMatrix, annot=True, vmax=1, square=True, cmap="Oranges")
plt.xlabel("互信息相关系数")
plt.show()
```
