```python
import pandas as pd
import datetime
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
from sklearn.datasets import load_iris
df_tmp = load_iris()
df = pd.DataFrame(df_tmp.data)
df.columns = df_tmp.feature_names
```

```python
plt.figure(figsize=(20, 8), dpi = 100)
hitmapData = df.corr(method = 'spearman') # method：可选值为{‘pearson’, ‘kendall’, ‘spearman’}
sns.heatmap(hitmapData, vmax=1, square=True, annot=True)

# plt.savefig('./figs/Correlation-Matrix.png')

plt.xticks(rotation = -70)
plt.yticks(rotation = 20)
plt.show()
```
