```python
import pandas as pd
import numpy as np
```

```python
file_path = r"附件1-秦皇岛港动力煤历史价格.xlsx"
df = pd.read_excel(file_path, header=1)
df["时间(该日期所在“周”的煤炭价格)年/月/日"] = pd.to_datetime(df["时间(该日期所在“周”的煤炭价格)年/月/日"])
df.set_index("时间(该日期所在“周”的煤炭价格)年/月/日", inplace=True)
df = df.resample('M').mean()
print(df)
```

```python
import matplotlib.pyplot as plt
data = df["价格平均值(元/吨)"].values
plt.plot(data)
```

```python
res_file = './tmp.xlsx'
df.to_excel(res_file)
```
