```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist
```

```python
file_path = r"C:\Users\Administrator\Desktop\2020\2020-51MCM-Problem A\附件1-秦皇岛港动力煤历史价格.xlsx"
df = pd.read_excel(file_path, header=1)
df.head(3)
```

```python
down = 0
up = 51
y1 = df["价格平均值(元/吨)"].tolist()[down:up]
y2 = df["价格低值(元/吨)"].to_list()[down:up]
y3 = df["价格高值(元/吨)"].to_list()[down:up]
df["时间(该日期所在“周”的煤炭价格)年/月/日"] = pd.to_datetime(df["时间(该日期所在“周”的煤炭价格)年/月/日"])
x = df["时间(该日期所在“周”的煤炭价格)年/月/日"][down:up]
x = [i.strftime("%Y%m%d") for i in x]
x.reverse()
```

```python
plt.figure(figsize=(20,8), dpi=80)

_x = range(len(x))

plt.plot(_x, y1, label='curve')

plt.title("curve", fontsize=20)
plt.xticks(_x, x, rotation=45)
plt.xlabel('year', fontdict={'fontsize':20})
plt.ylabel('quatity', fontdict={'fontsize':20})

plt.legend(loc='best', fontsize=20)
plt.show()
```
