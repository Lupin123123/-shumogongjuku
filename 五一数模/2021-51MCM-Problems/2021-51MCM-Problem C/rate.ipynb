```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"附件1(Appendix 1)2021-51MCM-Problem C.xlsx")
df
```

```python

for i in [1, 3, 4, 9, 10, 15]:
    bzc = df[i].std()
    aver = df[i].mean()

    # print(bzc)
    # print(aver)

    print("{},{}".format(aver-3*bzc, aver+2*bzc))
    ser = df[i]

    bool_id = ((ser.mean() - 3 * ser.std()) <= ser)  &   (ser <= (ser.mean() + 3 * ser.std()))
    idx = bool_id[bool_id==False].index.to_list()
    tmp = df.iloc[idx, i+1]
    
    tmp2 = tmp.to_numpy()

    tmp2 = abs(tmp2-aver)
    
    tmp2 = (tmp2-tmp2.min())/(tmp2.max()-tmp2.min())
    
    tmp3 = pd.DataFrame(index=idx, data=tmp2)
    
    tmp3.to_excel('rate{}.xlsx'.format(i))
    
```
