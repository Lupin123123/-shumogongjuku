```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tag1 = np.array([1, 1, 1, 0, 0, 1, 0, 0, 1, 1, 1, 0, 1, 1, 1])
num = np.arange(len(tag1))
P = np.array([0.67209, 0.75836,0.42741, 0.35476, 0.40207, 0.76641, 0.34897, 0.42207, 0.75485, 0.42431, 0.68043, 0.22613, 0.67488, 0.43989, 0.75952])
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(20,8), dpi=80)

plt.scatter(num[np.array(tag1==1)], P[np.array(tag1==1)], color='Red', s=300, label='SVM预测发生火灾')
plt.scatter(num[np.array(tag1==0)], P[np.array(tag1==0)], color='Green', s=300, label='SVM预测不发生火灾')
plt.plot(P, color='darkblue', linewidth=3, label='Logistic预测火灾发生概率')

plt.xticks(range(len(tag1)), fontsize=14)
plt.yticks(np.arange(0, 1.1, 0.1), fontsize=14)

plt.legend(fontsize=14)
plt.title("SVM和Logistic结果对比", fontsize=25)
plt.grid(axis='y')

plt.show()
```
