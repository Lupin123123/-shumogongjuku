```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('第三问作图.xlsx')
```

```python
y_label = df['四大指标'].to_list()
x_label = df.columns.to_list()[1:]
```

```python
x_label
```

```python
X = df.to_numpy()
X = X[:, 1:]
X = X.astype(np.float64)
```

```python
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10,8), dpi=80)

sns.heatmap(X, cmap='Oranges')

plt.xticks(np.arange(0,4)+0.5, x_label, rotation=0, fontsize=12)
plt.yticks(np.arange(0,len(y_label))+0.5, y_label)

plt.title("问题三各指标热力图", fontsize=16)
plt.show()
```
