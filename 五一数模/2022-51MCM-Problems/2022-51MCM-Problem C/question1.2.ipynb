```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel(r'附件1 报警数据集 (Attachment 1 Alarm data set).xlsx', header=1)
# df2 = pd.read_excel(r'附件2 故障数据集 (Attachment 2 Fault data set).xlsx', header=1) 
# df3 = pd.read_excel(r'question1_data.xlsx')
```

```python
df1.dropna(axis=0, inplace=True)
# df2.dropna(axis=0, inplace=True)
# df3.dropna(axis=0, inplace=True)
```

```python
lst_tmp1 = list(set(df1['隶属消防机构 (Subordinate to fire service)'].to_list()))
```

```python
for i in lst_tmp1:
    df_tmp = df1[df1['隶属消防机构 (Subordinate to fire service)'] == i]
    idx1 = df_tmp['是否属于误报 (Whether a misstatement)'] == '否，其中有1次为真实火灾 (False, one of them was a real fire)'
    df_tmp.loc[idx1==True, '火警次数 (Number of fire alarms)'] = 1

    idx2 = df_tmp['是否属于误报 (Whether a misstatement)'] == '是 (True)'
    df_tmp.loc[idx2==True, "火警次数 (Number of fire alarms)"] = 0

    df1_tmp = df_tmp.groupby(by='部件名称 (Component name)').sum()
    df1_tmp = pd.DataFrame(df1_tmp['火警次数 (Number of fire alarms)'])
    # print(df1_tmp)
    df1_tmp.to_excel('question1_zhenshi_count{}.xlsx'.format(i))
```
