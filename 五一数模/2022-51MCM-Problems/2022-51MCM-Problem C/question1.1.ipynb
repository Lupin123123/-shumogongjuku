```python
import numpy as np
import pandas as pd
import string
```

```python
df1 = pd.read_excel(r'附件1 报警数据集 (Attachment 1 Alarm data set).xlsx', header=1)
df2 = pd.read_excel(r'附件2 故障数据集 (Attachment 2 Fault data set).xlsx', header=1) 
```

```python
df1.dropna(axis=0, inplace=True)
df2.dropna(axis=0, inplace=True)
```

```python
lst_tmp1 = list(set(df1['部件名称 (Component name)'].to_list()))
lst_tmp2 = list(set(df2['部件名称 (Component name)'].to_list()))
```

```python
print(len((lst_tmp1)))
print(lst_tmp1)

X = pd.DataFrame(columns=lst_tmp1)
X['region'] = 0
```

```python
cnt = 0
for i in string.ascii_uppercase[:18]:
    path1 = 'question1_zongcishu_count{}大队 (Fire brigade {}).xlsx'.format(i,i)
    path2 = 'question1_zhenshi_count{}大队 (Fire brigade {}).xlsx'.format(i,i)

    df_tmp = pd.read_excel(path1)
    df_zhenshi = pd.read_excel(path2)

    lst_tmp = df_tmp['部件名称 (Component name)'].to_list()
    
    for j in range(len(lst_tmp)):
        X.loc[cnt, X.columns==lst_tmp[j]] = df_tmp.loc[j, '火警次数 (Number of fire alarms)']
        X.loc[cnt, 'region'] = ord(i)
        if (df_zhenshi.loc[j, '火警次数 (Number of fire alarms)']!=0): 
            X.loc[cnt, 'Y_hat'] = 1
        else:
            X.loc[cnt, 'Y_hat'] = 0
        # print(df_zhenshi.loc[j, '火警次数 (Number of fire alarms)'])
        cnt += 1
```

```python
X.fillna(0, inplace=True)
X.to_excel('X_Y.xlsx')
X
```
