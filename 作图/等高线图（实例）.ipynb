```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import Rbf
from metpy.interpolate import inverse_distance_to_grid
```

```python
df = pd.read_excel('地累积指数.xlsx')
```

```python
x = df['x(m)'].to_numpy()
y = df['y(m)'].to_numpy()
z = df['As (μg/g)'].to_numpy()
```

```python
step = 25
x_grid = np.arange(x.min(), x.max()+step, step)
y_grid = np.arange(y.min(), y.max()+step, step)
x_gridmesh, y_gridmesh = np.meshgrid(x_grid, y_grid)
```

```python
z_grid = inverse_distance_to_grid(x, y, z, x_gridmesh, y_gridmesh, r=3000)
plt.imshow(z_grid, origin='lower')
plt.show()
```
