```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'产品':['肉类','盐铁','纺织','木材']*2,
        '年份':[1046,1046,1046,1046,1047,1047,1047,1047],
       '诸侯':['秦','齐','楚','秦','秦','楚','齐','齐'],
       '产量':[180,140,300,200,150,60,80,320]}
 
df=pd.DataFrame(data, columns=['产品','年份','诸侯','产量'])

df_new = df
df_new['x'] = [1,2,3,4,5,6,7,8]
df_new['年份'] = [1,2,3,4,5,6,7,8]
df_new['产量'] = [10,20,13,24,15,60,70,80]
```

## plot预处理函数

```python
def plot():
    plt.figure(figsize=(6, 6), dpi=80)
    plt.rcParams["font.sans-serif"] = ["SimHei"]
    plt.rcParams["axes.unicode_minus"] = False
```

## subplots方法

```python
##绘图
fig, ax = plt.subplots(1, 2, figsize=(8, 4), dpi=80)

ax[0].plot(df_new['x'], df_new['年份'], color='green', linewidth=0.8)
ax[1].plot(df_new['x'], df_new['产量'], color='green', linewidth=0.8)
ax[0].grid()
ax[1].grid()

fig.suptitle("my plot title", fontsize=30)
fig.subplots_adjust(hspace=0.4, top=0.75)
 
plt.show()
```

## subplot方法

```python
import matplotlib.pyplot as plt
import numpy as np

def plot():
    fig = plt.figure(figsize=(20, 20), dpi=80)
    plt.rcParams["font.sans-serif"] = ["SimHei"]
    plt.rcParams["axes.unicode_minus"] = False

#plot 1:
x = np.array([0, 6])
y = np.array([0, 100])

plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("plot 1")

#plot 2:
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("plot 2")

#plot 3:
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("plot 3")

#plot 4:
x = np.array([1, 2, 3, 4])
y = np.array([4, 5, 6, 7])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("plot 4")

plt.suptitle("RUNOOB subplot Test", x=0.5, y=0.92, fontsize=30)
plt.show()
```
