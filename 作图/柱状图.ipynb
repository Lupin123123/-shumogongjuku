## 基于matplotlib绘制多组并列柱状图

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6), dpi=80)

size = 4

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

a = [0.55, 0.5, 0.5, 0.5]
b = [0.8324, 0.9346, 0.8192, 0.9563]

x = np.arange(size)                 # x轴坐标, size=5, 返回[0, 1, 2, 3, 4]

total_width, n = 0.8, 2             # 每种类型的柱状图宽度, 有a/b/c三种类型的数据，n设置为3
width = total_width / n
x = x - (total_width - width) / 2   # 重新设置x轴的坐标

plt.bar(x, a, width=width, label="相对成本")
plt.bar(x + width, b, width=width, label="完成率")


plt.xticks(x+0.2, ['1', '2', '3', '4'])  #设置x轴
plt.xlabel('编号')
plt.title('相对成本及完成率')
plt.legend()


# 写上数字
for i, j in zip(x, a):
    plt.text(i, j + 0.01, "%.2f" % j, ha="center", va="bottom", fontsize=7)
for i, j in zip(x + width, b):
    plt.text(i, j + 0.01, "%.2f" % j, ha="center", va="bottom", fontsize=7)

plt.savefig('问题二图片.jpg')

plt.show()
```

## 利用seaborn作图

```python
import pandas as pd
import numpy as np
 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
plt.figure(figsize=(10,5), dpi=80)

df = sns.load_dataset('tips')

ax = sns.barplot( data = df, x="day", y="tip", hue = 'sex')  # 通过DataFrame来指定
sns.move_legend(ax, loc='upper left')

plt.show()
```
