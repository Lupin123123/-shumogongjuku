```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
iris = sns.load_dataset('iris', data_home='.')
```

```python
iris
```

## 一图多绘

```python
fig, axes = plt.subplots(1, 1, figsize=(20, 6), dpi=100)
axesSub = sns.boxplot(x='species', y='sepal_length', data=iris, ax = axes)
axesSub = sns.stripplot(x='species', y='sepal_length', data=iris, ax=axesSub)  
# ax=axesSub，指定坐标性, 在之前箱型图基础上，继续绘制条带图
axesSub = sns.lineplot(x='species', y='sepal_length', data=iris, ax=axesSub)  
# ax=axesSub，指定坐标性, 在之前箱型图基础上，继续绘制折线图
```

## 多子图

```python
fig, ax =plt.subplots(1, 3, constrained_layout=True, figsize=(12, 3), dpi=100)

axesSub = sns.boxplot(x="species", y="sepal_length", data=iris, ax=ax[0])
axesSub.set_title('boxplot')
axesSub = sns.stripplot(x="species", y="sepal_length", data=iris, ax=ax[1])
axesSub.set_title('stripplot')
axesSub = sns.histplot(x="sepal_length", data=iris, ax=ax[2])
axesSub.set_title('histplot')

plt.suptitle('examplrs of subplots')
plt.show()
```

## PairGrid：成对关系网格子图

```python
plt.rcParams.update({'figure.dpi':100})
g = sns.PairGrid(iris, hue="species")
g = g.map(sns.scatterplot)
g = g.add_legend()  # 手动添加图例
```

```python
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], height=3)
g = g.map(sns.scatterplot, color='green')
```
